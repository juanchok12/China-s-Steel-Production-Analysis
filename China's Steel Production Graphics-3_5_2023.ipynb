```python

#Creating the treemap figure
fig_treemap_tot_16_21 = px.treemap(df_all_adv_steel_total, path=['Country'],
 values='Steel Output', 
 title='World Steel Output 2016-2022 (data source: World Steel Association)',
 hover_data={'Steel Output':True, 'Percentage of Total':False},
 template='plotly_dark',
 color=df_all_adv_steel_total['Country'],
 color_discrete_map={'China 52.54%':'#990000','Rest of world 12.58%':'#006666','Russia 3.93%':'#0000CC','Japan 5.34%':'#333300','South Korea 3.81%':'#CC0066','India 5.95%':'#660066','United States 4.49%':'#336699','Germany 2.19%':'#336600','Brazil 1.85%':'#800000','Turkey 1.98%':'#003399','Italy 1.26%':'#666699','United Kingdom 0.39%':'#003300','France 0.76%':'#CC00CC','Spain 0.72%':'#6600CC','Canada 0.69%':'#993366','Australia 5.79%':'#FFCC00','Mexico 1.02%':'#000066','Poland 0.49%':'#336699'})



#Costomize the hover tamplate to remove the 'id' and 'parent' columns, and adds the 'Steel Output' values in accounting format
fig_treemap_tot_16_21.update_traces(hovertemplate='<b>%{label}</b><br>Tons of steel: %{customdata[0]:,.0f}')
fig_treemap_tot_16_21.show()
```

```python
#Exporting tree map to plotly chart studio

py.plot(fig_treemap_tot_16_21, filename = 'treemap_tot_16_21', auto_open=True)

```

```python
#Create a dataframe form the 'csp 1917-2021-Updated-3-10-2023.xlsx' file knowing that the header row starts at row 4 and the data starts at row 5 and the data ends at row 21 and the data is in columns A:DC and the index column is column A
df_steel_linear = pd.read_excel('csp 1917-2021-Updated-3-10-2023.xlsx', sheet_name='Formatted Tables',header=3, index_col=0, usecols="A:DC",skiprows=0,nrows=17) 
df_steel_linear
```

```python
#Create a linear plot of the data in the dataframe df_steel_linear in which the x-axis is the time in years from 1917-2021 and the y-axis is the steel production in millions of tons for each country. Make sure that the legends are in the right side of the plot and that they are in the same order as the countries in the dataframe. Also, make sure that the plot is in the 'plotly_dark' template and that the plot is interactive. Make sure that I can the legand names also works as a filter in the plot.
#First, let's change the x and y axis
df_steel_transposed = df_steel_linear.transpose()

#Creation of the linear plot
fig_linear_steel_1917_2022 = px.line(df_steel_transposed, 
                     x=df_steel_transposed.index, 
                     y=df_steel_transposed.columns, 
                     title='World Steel Production 1917-2022 (data source: World Steel Association)', 
                     template='plotly_dark', 
                     #hover_data={'value':True, 'variable':False}, 
                     labels={'value':'Millions of Tons', 'variable':'Country'})
                     #color_discrete_map={'China':'#990000','Rest of world':'#006666','Russia':'#0000CC','Japan':'#333300','South Korea':'#CC0066','India':'#660066','United States':'#336699','Germany':'#336600','Brazil':'#800000','Turkey':'#003399','Italy':'#666699','United Kingdom':'#003300','France':'#CC00CC','Spain':'#6600CC','Canada':'#993366','Australia':'#FFCC00','Mexico':'#000066','Poland':'#336699',})
#Display
fig_linear_steel_1917_2022
```

```python
#Exporting linear plot to plotly chart studio
py.plot(fig_linear_steel_1917_2022, filename='linear_plot_steel_1917_2022', auto_open=True)
```

```python
#Measuring the exponential growth rate of China's steel production from 1949-2022
#First, let's create a dataframe with the data from 1949-2022
df_steel_china_measurement = df_steel_transposed.loc['1949':'2022','China']

#Now, let's calculate the exponential growth rate
df_steel_china_measurement['Exponential Growth Rate'] = df_steel_china_measurement['1949':'2022'].pct_change()

#Find the average exponential growth rate
df_steel_china_measurement['Exponential Growth Rate'].mean()













```
